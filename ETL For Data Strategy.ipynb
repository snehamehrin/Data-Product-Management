```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile

df = pd.read_csv('Flyber_event_logs.csv')

print("Column headings:")
print(df.columns)
```

```python
#Taking a first look at the daa
df.head()
```

```python
df.drop(columns={'Unnamed: 0'},inplace=True)
df['date']=df['event_time'].dt.date
```

```python
#Question 1 : Find number of events grouped by date
df_transposed=pd.DataFrame(df.groupby('date')['event_uuid'].count())
df_transposed.T
```

```python
#Question 2 : Find Number of events grouped by date and event_type
df_pivot=df.pivot_table(index='event_type',columns='date',values='event_uuid',aggfunc='count')
df_pivot
```

```python
#Question 3 : Find Number of events grouped by date and device_type
df_pivot_car=df.pivot_table(index='device_type',columns='date',values='event_uuid',aggfunc='count')
df_pivot_car
```

```python
#Question 4 : Find Number of events grouped by event_page and event_type
df_pivot_page=df.pivot_table(index='event_page',columns='date',values='event_uuid',aggfunc='count')
df_pivot_page
```

```python
#Question 5 : Find Number of events grouped by date and location
df_pivot_location=df.pivot_table(index='user_neighborhood',columns='date',values='event_uuid',aggfunc='count')
df_pivot_location
```
